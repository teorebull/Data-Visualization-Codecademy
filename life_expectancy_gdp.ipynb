## GDP and Life Expectancy relationship

The goal of this project is to find relationships between the GDP in US dollars and the life expectancy of 6 different countries.
We'll mostly focus on the visualization and derive conclusions from there 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all_data.csv')
df.head()
```

```python
# First inspection to make sure there aren't any null values
print(df.isna().sum())
# Analyzing the type of each feature
df.dtypes
```

```python
# Find out which countries we have
print(df['Country'].value_counts())
# Prepare the data for each subplot
chile = df[df['Country'] == 'Chile']
china = df[df['Country'] == 'China']
germany = df[df['Country'] == 'Germany']
mexico = df[df['Country'] == 'Mexico']
usa = df[df['Country'] == 'United States of America']
zimbabwe = df[df['Country'] == 'Zimbabwe']

# Create the labels for the legend
country_set = set()
for country in df['Country']:
    country_set.add(country)
country_list = list(country_set)
country_list.sort()
print(country_list)

# Create visualization
fig = plt.figure(figsize=(7,5))
ax = fig.subplots(2,3)
fig.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.75, hspace=0.75)

# Chile subplot
plt.subplot(2,3,1)
plt.xlabel("Life Expectancy at birth (years)")
plt.ylabel("GDP in Trillions of US dollars")
plt.title("Country = Chile")
plt.scatter(x='Life expectancy at birth (years)', y='GDP', data=chile, color='red')
# plt.legend([country_list[0]], loc="lower right")

# China subplot
plt.subplot(2,3,2)
plt.xlabel("Life Expectancy at birth (years)")
plt.ylabel("GDP in Trillions of US dollars")
plt.title("Country = China")
plt.scatter(x='Life expectancy at birth (years)', y='GDP', data=china, color='green')
# plt.legend([country_list[1]], loc="lower right")

# Germany subplot
plt.subplot(2,3,3)
plt.xlabel("Life Expectancy at birth (years)")
plt.ylabel("GDP in Trillions of US dollars")
plt.title("Country = Germany")
plt.scatter(x='Life expectancy at birth (years)', y='GDP', data=germany, color='black')
# plt.legend([country_list[2]], loc="lower right")

# Mexico subplot
plt.subplot(2,3,4)
plt.xlabel("Life Expectancy at birth (years)")
plt.ylabel("GDP in Trillions of US dollars")
plt.title("Country = Mexico")
plt.scatter(x='Life expectancy at birth (years)', y='GDP', data=mexico, color='brown')
# plt.legend([country_list[3]], loc="lower right")

# USA subplot
plt.subplot(2,3,5)
plt.xlabel("Life Expectancy at birth (years)")
plt.ylabel("GDP in Trillions of US dollars")
plt.title("Country = USA")
plt.scatter(x='Life expectancy at birth (years)', y='GDP', data=usa, color='blue')
# plt.legend([country_list[4]], loc="lower right")

# Zimbabwe subplot
plt.subplot(2,3,6)
plt.xlabel("Life Expectancy at birth (years)")
plt.ylabel("GDP in Trillions of US dollars")
plt.title("Country = Zimbabwe")
plt.scatter(x='Life expectancy at birth (years)', y='GDP', data=zimbabwe, color='yellow')
# plt.legend([country_list[5]], loc="lower right")

fig.legend([plt.subplot(2,3,1), plt.subplot(2,3,2), plt.subplot(2,3,3), plt.subplot(2,3,4), plt.subplot(2,3,5), plt.subplot(2,3,6)], 
           labels=country_list,bbox_to_anchor=(1.3, 1), fontsize=7, loc="upper right", title="Country") 
plt.show()
plt.clf()
```
